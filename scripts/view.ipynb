```python
import pandas as pd

df = pd.read_pickle("cold_storage/NSE Main Pickle.pkl")
df.head()
```

```python
df['timestamp'] = df['Date']
df['o'] = df['Open (Unit Curr)']
df['h'] = df['High (Unit Curr)']
df['l'] = df['Low (Unit Curr)']
df['c'] = df['Close (Unit Curr)']
df['v'] = df["Volume (000's)"] * 1000
df['v'].fillna(0, inplace=True)
df['v'] = df['v'].astype(int)
df.drop(columns=['Open (Unit Curr)', 'Close (Unit Curr)', 'High (Unit Curr)', 'Low (Unit Curr)', "Volume (000's)", "Value", 'Index', 'Date'], inplace=True)
df.head()
```

```python
for col in list(df.columns):
    if col != 'timestamp' and df[col].dtype != 'int':
        print(df[col].dtype)
        df[col] = df[col].astype(float)
        print(df[col].dtype)
```
